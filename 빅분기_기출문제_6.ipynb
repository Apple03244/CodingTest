## 작업유형 1

DataUrl = https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e6_p1_1.csv

데이터 출처 : https://bigdata-119.kr/

데이터 설명 : 소방신고별 신고시각과 출동시각 데이터 컬럼 설명 : 신고일자 (yyyymmdd) / 신고시각 (HHMMSS , 00시 , 00분의 경우 등 생략)

```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e6_p1_1.csv')
```

각 구급 보고서 별 출동시각과 신고시각의 차이를 ‘소요시간’ 컬럼을 만들고 초(sec)단위로 구하고 소방서명 별 소요시간의 평균을 오름차순으로 정렬 했을때 3번째로 작은 소요시간의 값과 소방서명을 출력하라

```python
data
```

```python
data['소요시각'] = (
                pd.to_datetime(
                    data['출동일자'].astype('str') + data['출동시각'].astype('str').str.zfill(6)
                )

                -

                pd.to_datetime(
                    data['신고일자'].astype('str') + data['신고시각'].astype('str').str.zfill(6)
                )
                ).dt.total_seconds()


result = data.groupby(['소방서명'])['소요시각'].mean().sort_values().reset_index()
result
```

```python
data.info()
```

```python
pre_data=data[data.columns.difference(["신고일자","신고시각","출동일자","출동시각"])].copy()
# 컬럼 타입 변환
def pluszero(x):
  x="0"*(6-len(x))+x
  timeletter=[x[i:i+2] for i in range(0,len(x),2)]
  return ":".join(timeletter)
  #return ":".join([x[i,i+2] for i in range(0,len(x),2)])
for col in ["신고시각","출동시각"]:
  data[col]=data[col].astype("str")
  data[col]=data[col].apply(pluszero)
```

```python
data.head(3)
```

```python
pre_data["출동시간"]=data["출동일자"].astype("str")+" "+data["출동시각"]
pre_data["신고시간"]=data["신고일자"].astype("str")+" "+data["신고시각"]
```

```python
pre_data["출동시간"]=pd.to_datetime(pre_data["출동시간"])
pre_data["신고시간"]=pd.to_datetime(pre_data["신고시간"])
```

```python
pre_data["소요시간"]=pre_data["출동시간"]-pre_data["신고시간"]
```

```python
result_df=pre_data[["소방서명","소요시간"]]
result_df["소요시간"]=result_df["소요시간"].apply(lambda x:x.total_seconds())
print(pd.DataFrame(result_df.groupby(["소방서명"])["소요시간"].mean()).sort_values("소요시간").iloc[2])
```

## 작업유형 1-3

DataUrl = https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e6_p1_3.csv

5대범죄 발생현황 : https://www.data.go.kr/data/15085721/fileData.do#layer_data_infomation & 후처리

```python
data3 = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e6_p1_3.csv')
data3.head(3)
```

**5대 범죄(절도, 사기, 배임, 방화, 폭행)의 월별 총 발생건수를 총범죄수라고 표현하자. 18,19년의 각각 분기별 총범죄수의 월평균 값을 구했을때 최대값을 가지는 년도와 분기를 구하고 해당 분기의 최댓값의 사기가 발생한 월의 사기 발생건수를 출력하라(1분기:1,2,3월 / 2분기 : 4,5,6월 / 3분기 7,8,9월 / 4분기 10,11,12월 , 1분기 월평균 : 1,2,3월의 총범죄수 평균) **

```python
import datetime as dt
```

```python
def trans(x):
  x=x.replace("년","")
  x=x.replace("월","")
  x=x.replace("_","/")
  return x
```

```python
pre_data=data3[data3.columns.difference(["범죄분류"])].copy()
pre_data["일자"]=pd.to_datetime(data3["범죄분류"].apply(trans))
```

```python
pre_data["총범죄수"]=0
```

```python
pre_data["총범죄수"]=pre_data[pre_data.columns.difference(["일자"])].sum(axis=1)
```

```python
pre_data["년도"]=pre_data["일자"].apply(lambda x:x.year)
pre_data["분기"]=pre_data["일자"].apply(lambda x:int((x.month-1)//3)+1)
```

```python
result=pd.DataFrame(pre_data.groupby(["년도","분기"])["총범죄수"].mean()).sort_values("총범죄수",ascending=False).reset_index().iloc[0]
```

```python
year=result["년도"]
quater=result["분기"]
```

```python
idx1=pre_data["년도"]==year
idx2=pre_data["분기"]==quater
```

```python
result_data=pre_data[idx1&idx2]
result_data
```

```python
result_data.sort_values(["사기"],ascending=False).iloc[0]["사기"]
```

## 작업유형 3

A 도시의 남성 600명과 여성 550명이 있다. 남성들 중 흡연자 비율은 0.2이며 여성들 중 흡연자 비율은 0.26이다.
남성과 여성 간에 흡연 여부에 따른 인구 비율이 다른지 확인하고 싶다. 유의 수준 0.05하 귀무가설에 대해 기각 / 채택 여부와 p-value값을 각각 출력하라

```python
data3=pd.crosstab(index=["M","W"],columns=["S","N"])
```

```python
data3.loc["M"]["S"]=600*0.2
data3.loc["M"]["N"]=600*0.8
data3.loc["W"]["S"]=550*0.26
data3.loc["W"]["N"]=550*0.74
```

```python
data3
```

```python
import scipy as sp
sp.stats.chi2_contingency(data3)
```
