```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e6_p1_1.csv')
```

각 구급 보고서 별 출동시각과 신고시각의 차이를 ‘소요시간’ 컬럼을 만들고 초(sec)단위로 구하고 소방서명 별 소요시간의 평균을 오름차순으로 정렬 했을때 3번째로 작은 소요시간의 값과 소방서명을 출력하라

```python
data
```

```python
data.info()
```

```python
import datetime as dt

def timesplit(x):
  x=str(x)
  if len(x)!=6:
    x="0"*(6-len(x))+x
  return dt.datetime.strptime(":".join([x[i:i+2] for i in range(0,len(x),2)])
  ,"%H:%M:%S")
```

```python
result_data=data.copy()
result_data["출동시각"]=data["출동시각"].apply(timesplit)
result_data["신고시각"]=data["신고시각"].apply(timesplit)
```

```python
result_data["신고일자"]=result_data["신고일자"].apply(lambda x:pd.to_datetime(str(x)))
result_data["출동일자"]=result_data["출동일자"].apply(lambda x:pd.to_datetime(str(x)))
```

```python
result_data.소요시간=result_data.소요시간.apply(lambda x:x.total_seconds())
```
