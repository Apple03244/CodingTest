```python
#여행(왜 저랩..)
def solution(N,s):
    move=dict(zip(['L','R','U','D'],[-1,1,-1,1]))
    s=list(s)
    positoin={
        "r":[["R","L"],1],
        "c":[["U","D"],1]
    }
    for x in s:
        for y in positoin:
            if x in positoin[y][0]:
                if (0<positoin[y][-1]+move[x]<=N):
                    positoin[y][-1]+=move[x]
    return (positoin["c"][-1],positoin["r"][-1])
```

```python
#시간
def solution(x):
    def trans(num):
        return str(num) if len(str(num))==2 else "0"+str(num)
    #자리
    time={"H":list(map(trans,range(x+1))),
          "MS":list(map(trans,range(60)))}
    result=0
    for h in time["H"]:
        for m in time["MS"]:
            for s in time["MS"]:
                if str(x) in h+m+s:
                    result+=1
    return result
    
    
```

```python
import numpy as np
#게임 개발
def gamestart():
    start=input("게임을 시작하시겠습니까? (Y/N) ")
    if start.upper()=="Y": 
        try:   
            n,m=map(int,input("맵의 크기를 입력하세요 : ").split(" "))
            x,y,z=map(int,input("캐릭터의 위치와 방향을 입력하세요 : ").split(" "))
            map_dict={} #껍질
            zero=[]
            for j in range(m):
                map_dict[j]=list(map(lambda x: 0 if not int(x) else 1,input(f"{j}번쨰 행({n}개)을 입력하세요 : ").split(" ")))[0:n]
                for k in range(n):
                    if map_dict[j][k]==0:
                        zero.append([j+1,k+1])
            result=[[x,y]]
            while True:
                for location in result:
                    for area in zero:
                        if (abs(location[0]-area[0])+abs(location[1]-area[1]))<=1:
                            result.append(area)
                            zero.remove(area)
                    pass
            return 
        except:
            print("마지막 입력시 SPACE바를 누르지 마세요^^")
    else:
        print("다시 실행하세요")
```
